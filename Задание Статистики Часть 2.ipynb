# Центральная предельная теорема на примере распределения хи-квадрат

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

### Зададим количество степеней свободы k = 5

```python
# сгенерируем выборку объема 1000
chi2_rv = sts.chi2(5)

sample = chi2_rv.rvs(1000)
```

```python
# выведем значения выборки
print sample
```

```python
x = np.linspace(1,10,1000)# определим массив точек для построения графика плотности распределения
pdf = chi2_rv.pdf(x)# вычислим значения теоретической плотности распределения
```

```python
# построим гистограмму выборки и теоретическую плотность распределения
plt.plot(x, pdf, label='theoretical PDF')
plt.hist(sample, bins =60, range=(0,30), normed=True)
plt.ylabel('samples number')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
# Для распределения хи-квадрат: среднее (мат.ожидание) Е = k, дисперсия D = 2*k
k=5 # количество степеней свободы
E=k # среднее(мат.ожидание)
D=2*k # дисперсия
print E
print D
```

### n=5

```python
n=5
sample=np.array([])#создадим массив выборочных средних
i=1
while i <= 1000:# зададим цикл генерации случайных выборок количества n
    samplei = chi2_rv.rvs(n)
    meani=sum(samplei)/n
    sample=np.append(sample,meani)
    i+=1
#определим параметры нормального распределения для массива 1000 выборочных средних при n=5 с помощью теоретических параметров
E=5.0
D=10.0
Dn=D/n # дисперсия для нормального распределения приближающего выборку средних
sigma=math.sqrt(Dn)# стандартное отклонение - квадратный корень из дисперсии
norm = sts.norm(E, sigma)# определяем нормальное распределение
pdf = norm.pdf(x)# получаем значения плотности распределения
# построим гистограмму массива 1000 выборочных средних при n=5 и график плотности соответствующего нормального распределения
plt.plot(x, pdf, label='PDF(n=5)')
plt.hist(sample, bins =70, range=(0,30), normed=True)
plt.ylabel('samples number')
plt.xlabel('$x$')
plt.legend(loc='best')
```
